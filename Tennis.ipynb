# Run Multi-Agent DDPG

```python
from unityagents import UnityEnvironment

env = UnityEnvironment(file_name="Tennis_Windows_x86_64/Tennis_Windows_x86_64/Tennis.exe", no_graphics=True)
```

```python
from maddpg import MADDPG
from run import ddpg_multi_agent
import numpy as np

scores = []
ddpg_multi = MADDPG(env)
```

```python
scores += ddpg_multi_agent(env, ddpg_multi, n_episodes=10000, save_every=100)
```

## Reward Plot

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10, 5))

plt.plot(scores, color='red', alpha=0.2, label='Multi-Agent DDPG')
plt.plot(pd.Series(scores).rolling(100).mean(), color='k', label='Mean Score')
plt.hlines(0.5, 0, len(scores), color='purple', linestyle='--', label='Target Score')
plt.xlabel('Episode')
plt.ylabel('Score')
plt.legend()
ax = plt.gca()

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
```
